# ** Задача №1 (стр. 172)**

    Построить окружность, определив координаты ее центра и радиус:
            x^2-4x+y^2+6y=12
            
            Решение:
             Приведем уравнение к кононическому виду
            
            ((x^2)-4x+4)-4+((y^2)+6y+9)-9-12=0
            
            (x-2)^2+(y+3)^2-25=0
            
            (x-2)^2+(y+3)^2=25
             
           Перейдя к кононическому виду уравнения получаем
             Центр окружности проходит через точку с координатами (2;-3) и радиуса R=5

```
import pylab
import matplotlib.patches
import matplotlib.lines
import matplotlib.path


def drawLine (axes):
   # Радиус
  
    x0 = 2
    y0 = -3
    x1 = 2
    y1 = 2

    line = matplotlib.lines.Line2D ([x0, x1], [y0, y1], color="k")
    axes.add_line (line)

    pylab.text (0.1, 0.1, "R=5", horizontalalignment="left")
    
def drawCircles (axes):
    #Окружность
   
    circle1 = matplotlib.patches.Circle((2, -3),  radius=0.1,  color="r")
    axes.add_patch  (circle1)
    pylab.text (2.3,- 3.1, "(-2;3)", horizontalalignment="left")

    circle2 = matplotlib.patches.Circle((2, -3), radius=5, fill=False, color="r")
    axes.add_patch  (circle2)
  #  pylab.text (2, -3, "Circle", horizontalalignment="center")


  

if __name__ == "__main__":
    pylab.xlim (-10,10)
    pylab.ylim (-10, 10)
    pylab.grid()

    #Текущие оси        

    axes = pylab.gca()
    axes.set_aspect("equal")

    drawLine (axes)
    drawCircles (axes)
      
    pylab.show()
```

# ** Задача №3 (стр. 241)**

          Построить кривую
          
          (x-2)^2+(y+3)^2=16
          
          Решение
             
             Центр данной окружности проходит через точку с координатами (2;-3) и радиуса R=4



```
import pylab
import matplotlib.patches
import matplotlib.lines
import matplotlib.path


def drawLine (axes):
   # Радиус
  
    x0 = 2
    y0 = -3
    x1 = 2
    y1 = 1
    line = matplotlib.lines.Line2D ([x0, x1], [y0, y1], color="k")
    axes.add_line (line)

    pylab.text (0.1, -0.7, "R=4", horizontalalignment="left")
    
def drawCircles (axes):
    #Окружность
   
    circle1 = matplotlib.patches.Circle((2, -3),  radius=0.1,  color="r")
    axes.add_patch  (circle1)
    pylab.text (2.3,- 3.1, "(-2;3)", horizontalalignment="left")

    circle2 = matplotlib.patches.Circle((2, -3), radius=4, fill=False, color="r")
    axes.add_patch  (circle2)
  #  pylab.text (2, -3, "Circle", horizontalalignment="center")


  

if __name__ == "__main__":
    pylab.xlim (-10,10)
    pylab.ylim (-10, 10)
    pylab.grid()

    #Текущие оси        

    axes = pylab.gca()
    axes.set_aspect("equal")

    drawLine (axes)
    drawCircles (axes)
      
    pylab.show()
```

# ** Задача №4 (стр. 252)**

      Привести уравнения кривых к каноническому виду и построить эти кривые.
      
                  x^2+4y^2-2x-3=0
                  
             Решение
                      Данное уравнение имеет следующию каноническую форму ((x-1)^2)/4+(y)^2/1=1
                      Полученное уравнение, является уравнением элипса, где (1;0)-центр элипса, a=2,b=1 - 
                      фокусы элипса



```
from matplotlib.patches import Ellipse
import matplotlib as mpl
%matplotlib inline
from matplotlib import pyplot as plt
from math import sqrt, exp

#Центр элипса
mean = [ 1 ,  0]
#Полуоси
width = 2+2
height = 1+1

ell = mpl.patches.Ellipse(xy=mean, width=width, height=height)
fig, ax = plt.subplots()

ax.add_patch(ell)
ax.set_aspect('auto')
ax.autoscale()
plt.show()
```

# ** Задача №5 (стр. 279)**

    Определить вид и построить заданные поверхности:
            
            ((x^2)/4)-((y^2)/9)+z^2=1
            
       Решение
       
            Данное уравнение описывает гиперболойд, где а=2, b=3, с=1
      



```
from __future__ import division

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=plt.figaspect(1))  # Square figure
ax = fig.add_subplot(111, projection='3d')

r=1;
u=np.linspace(-2,2,200);
v=np.linspace(0,2*np.pi,60);
[u,v]=np.meshgrid(u,v);

a = 2
b = 3
c = 1

x = a*np.cosh(u)*np.cos(v)
y = b*np.cosh(u)*np.sin(v)
z = c*np.sinh(u)

ax.plot_surface(x, y, z,  rstride=5, cstride=5, color='b')

plt.show()
```
